# <center> numpy application: planet hunting with radial velocity observations </center>

## Background

Planets orbiting their host stars will orbit a common center of mass. The planet orbits in an ellipse, but the star will mirror this orbit in a smaller ellipse. Observing the star along our line of sight, we can potentially detect a component of this <i>wobbling</i> motion, in the form of a Doppler shift in its spectrum. This Doppler shift can be translated into a velocity, or <b>radial velocity</b>. We can infer the presence of planets orbiting other stars by taking snapshots of the star's radial velocity as a function of time.

The task of converting a stellar spectrum to a Doppler velocity is not trivial. In short, one way is to split a spectrum into roughly 700 <b>chunks</b> and compute a velocity for each chunk. A statistical weight is also assigned to each chunk depending on the quality of the Doppler content (e.g., number of spectral lines). Low Doppler content is given relatively low or zero weight, whereas high Doppler is given greater weight. In the end, the radial velocity of a particular spectrum is the <i>weighted average</i> of these velocities.

## Goals of this demo

There are two tasks we will accomplish here:
1. Compute these weighted average radial velocities for a real star.
2. Search the radial velocity time series for significant periodic signals that may correspond to an orbiting exoplanet.

The whole analysis should only take a couple seconds to run.

So let's load up numpy and some plotting utilities.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

This function simply converts an ascii file to a 2d numpy array.

```python
def file_to_array(filename):
    with open(filename) as f:
        array  = f.read().splitlines()
    
    return np.array([ [float(x) for x in i.split()] for i in array ])
```

We're going to read in two files. Each row corresponds to a particular spectrum/observation of HD 9407, a quiet nearby G dwarf. Each column corresponds to information regarding a particular spectral chunk. For this particular exercise, there are 628 spectra and 718 chunks per spectrum.

- 9407_rv.ascii is a 2d array of velocity information.
- 9407_wt.ascii is a 2d array of statistical weight information.

So let's read in the files and make sure that the number of spectra and chunks match between the files.

```python
vels_data = file_to_array('9407_rv.ascii')
wts_data  = file_to_array('9407_wt.ascii')

assert vels_data.shape == wts_data.shape
```

If the above line gave you no errors, then we can see how many chunks ($\tt{nChunks}$) and observations ($\tt{nObs}$) there are. 

```python
nObs, nChunks = vels_data.shape
print("Number of spectral chunks:", nChunks)
print("Number of observations:", nObs)
```

Now, let's compute the weighted averaged radial velocity for each spectrum. For the $i$th velocity $<v>_i$, this is computed with...
$$<v>_i = \frac{\sum_{j=1}^{\tt{nChunks}} w_{i,j} v_{i,j}}{\sum_{j=1}^{\tt{nChunks}} w_{i,j} } $$

...where $w_{i,j}$ and $v_{i,j}$ are the weight and velocity respectively of the $i$th observation and $j$th chunk.

This calculation could be done in one line, but I will split the operations in the numerator and denominator to show the steps.

```python
# Compute the sum of weights (denominator)
denominators = np.array( [sum(line) for line in wts_data] )

# Compute the outer product of weights and velocities (numerator)
numerators = np.sum( vels_data * wts_data, axis=1 )

# Compute all rvs
rvs = numerators / denominators
```

Part 1 is done!

Now onto Part 2. Let's see what the radial velocities look like as a function of time. We'll need to download the list of observation times.

```python
with open('9407_times.ascii') as f:
    times = np.array(list(map(float, f.read().splitlines())))
```

And plot the data. For the sake of this exercise, I don't include uncertainties in the RV measurements. 

```python
plt.plot(times, rvs, 'o')
plt.xlabel('Time - JD 2,440,000 [days]', fontsize=16)
plt.ylabel('Radial Velocity [m/s]', fontsize=16)
plt.show()
```

We can compute a Lomb-Scargle periodogram to get assess the prominent periodicities in this unevenly sampled time series. There are several implementation of this in Python. Jake Vanderplas provides a nice overview of them on [his blog](https://jakevdp.github.io/blog/2015/06/13/lomb-scargle-in-python/). I will use one from scipy since it comes with the standard Anaconda installation and works with data without uncertainties. A lot of this is just copied from the blog post.

```python
from scipy.signal import lombscargle

# Choose a period grid
pmin, pmax = 0.1, 6000.
periods = np.logspace(np.log10(pmin), np.log10(pmax), 2000)
ang_freqs = 2 * np.pi / periods

# Compute the (unnormalized) periodogram
# Note pre-centering of y values!
power = lombscargle(times, rvs - rvs.mean(), ang_freqs)

# Normalize the power
N = len(times)
power *= 2 / (N * rvs.std() ** 2)

# Plot the results
plt.semilogx(periods, power)
plt.xlabel('Period [days]', fontsize=16)
plt.ylabel('Lomb-Scargle Power', fontsize=16)
plt.show()
```

The strongest signals here correspond to a yearly/daily aliases and various harmonics. To really assess the significance of these signals is left for another discussion.
